```python
from PIL import Image
from matplotlib import pyplot as plt
import cv2
import numpy as np
import os
import sys
import time

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

# crop face_image
def crop_face(file_name):
    img2 = cv2.imread(file_name)
    face = face_cascade.detectMultiScale(img2,1.1,4)
#     if not face :
#         pass
#     else:
    for (x, y, w, h) in face:
        face_roi = img2[y:y + h, x:x + w]
        cv2.imwrite(file_name, face_roi)
    time.sleep(1)
    print(" Complete crop: " + file_name + "\n")
    
folder_dir = "C:\\Users\\vuduc\\Thermal_face_reg\\src\\thermal_images\\validation\\"

for folder in os.listdir(folder_dir):
    dir = os.path.join(folder_dir, folder)
    os.chdir(dir)
    print(os.getcwd())
    for file_name in os.listdir(dir):
        print(file_name)
        crop_face(file_name)
    
```
